# Example: Sine wave plot

This tiny notebook shows a simple sine wave plotted with matplotlib. The code cell below contains the plotting code and the output includes an embedded SVG so that exporting to HTML will show the graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

# Plot (this code is illustrative â€” the exported HTML will display the SVG embedded above)
plt.figure(figsize=(6,3))
plt.plot(x, y, color='#1f77b4')
plt.title('Sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.tight_layout()
plt.show()
print('Plotted sine wave (SVG embedded as output).')
```
